### Ken Wilbur Target Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew
```

```python
# Load data and print shape
# Load the data
df = pd.read_csv('train.csv')
print(df.shape)
```

```python
# Extract the target data from the last column
target = df.iloc[:, -1]

mean = np.mean(target)
median = np.median(target)
std_dev = np.std(target)
data_range = np.ptp(target)
target_min = np.min(target)
target_max = np.max(target)
target_skewness = skew(target)

print(f"Mean of target: {mean :.3f}")
print(f"Median of target: {median :.3f}")
print(f"Standard Deviation of target: {std_dev :.3f}")
print(f"Range of target: {data_range :.3f}")
print(f"Minimum of target: {target_min :.3f}")
print(f"Maximum of the target: {target_max :.3f}")
print(f"Skewness of target: {target_skewness :.3f}")
```

```python
# Kernel Density Estimate plot
plt.figure(figsize=(8, 6))
target.plot(kind='kde')
plt.title('Kernel Density Estimate of Target Values')
plt.xlabel('Target Value')
plt.ylabel('Density')
plt.show()
```

### Target Data Exploration Findings 
Mean of target: $0.867$<br>

Range of target: $3.500$<br>
Minimum of target: $0.000$<br>
Maximum of the target: $3.500$<br>
Skewness of target: $1.326$<br>
### Target Data KDE Interpretation
The Kernel Density Estimate distribution suggests the target data is skewed to the right but the mode of the target data is actually less than the mean.
